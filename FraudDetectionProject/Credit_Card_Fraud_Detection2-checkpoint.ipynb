```python
import pandas as pd



df = pd.read_csv('creditcard.csv')
print(df.shape)
df.head()

     
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Class distribution
sns.countplot(x='Class', data=df)
plt.title("Class Distribution")

# Correlation
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title("Feature Correlation Matrix")

     
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
import numpy as np

# Feature Engineering
df['Hour'] = (df['Time'] // 3600) % 24
df['LogAmount'] = np.log1p(df['Amount'])

# Use only relevant features
X = df.drop(['Class', 'Time', 'Amount'], axis=1)
y = df['Class']

over = SMOTE(sampling_strategy=0.1, random_state=42)
under = RandomUnderSampler(sampling_strategy=0.5, random_state=42)

resample_pipeline = Pipeline(steps=[('o', over), ('u', under)])
X_res, y_res = resample_pipeline.fit_resample(X, y)
```

```python

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import RobustScaler
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline as ImbPipeline
     
```

```python
models = {
    'LogisticRegression': {
        'model': LogisticRegression(solver='liblinear'),
        'params': {
            'clf__C': [0.01, 0.1, 1, 10]
        }
    },
    'DecisionTree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'clf__max_depth': [3, 5, 10],
            'clf__min_samples_split': [2, 5, 10]
        }
    },
    'XGBoost': {
        'model': XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
        'params': {
            'clf__n_estimators': [100, 200],
            'clf__max_depth': [3, 5, 7],
            'clf__learning_rate': [0.01, 0.1]
        }
    }
}
     
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)



print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
     
```

```python

for name, config in models.items():
    print(f"\n🔍 Training and tuning model: {name}")

    model_pipeline = ImbPipeline([
        ('scaler', RobustScaler()),
        ('clf', config['model'])
    ])

    grid = GridSearchCV(
        estimator=model_pipeline,
        param_grid=config['params'],
        scoring='average_precision',
        cv=3,
        n_jobs=-1,
        verbose=1
    )

    grid.fit(X_train, y_train) # Fit on training data

    # Predict on the testing data that has the engineered features
    y_pred = grid.predict(X_test)
    y_pred_proba = grid.predict_proba(X_test)[:, 1] # Get probabilities for the positive class

    print(f"✅ Best Parameters for {name}: {grid.best_params_}")
    print(classification_report(y_test, y_pred))
     
```

```python
# prompt: Evaluate using F1 score, Precision-Recall, AUC-ROC.

from sklearn.metrics import average_precision_score, roc_auc_score, precision_recall_curve, roc_curve, auc

# Continue from the loop where you evaluate each model
for name, config in models.items():
    print(f"\n🔍 Training and tuning model: {name}")

    model_pipeline = ImbPipeline([
        ('scaler', RobustScaler()),
        ('clf', config['model'])
    ])

    grid = GridSearchCV(
        estimator=model_pipeline,
        param_grid=config['params'],
        scoring='average_precision',  # Keep AP for grid search if that's desired for tuning
        cv=3,
        n_jobs=-1,
        verbose=1
    )

    # Fit the model on the resampled training data
    grid.fit(X_train, y_train)

    # Predict on the original test set (or resampled test set, depending on evaluation goal)
    # If evaluating on the original scale, predict on X_test derived from original data before resampling
    # If evaluating the performance on the resampled distribution, use X_test derived from X_res
    # Assuming you want to evaluate the model trained on resampled data on the resampled test set:
    y_pred = grid.predict(X_test)
    y_pred_proba = grid.predict_proba(X_test)[:, 1] # Get probabilities for the positive class


    print(f"✅ Best Parameters for {name}: {grid.best_params_}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Evaluate using F1-score (already in classification_report, but explicit)
    # F1-score is calculated and printed in the classification_report above.

    # Evaluate using Precision-Recall AUC
    pr_auc = average_precision_score(y_test, y_pred_proba)
    print(f"\nPrecision-Recall AUC for {name}: {pr_auc:.4f}")

    # Evaluate using ROC AUC
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    print(f"ROC AUC for {name}: {roc_auc:.4f}")

    # Optional: Plot Precision-Recall curve
    precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, label=f'{name} (AP = {pr_auc:.4f})')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Optional: Plot ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    roc_auc_plot = auc(fpr, tpr)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc_plot:.4f})')
    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import joblib
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler

# Optional: Use best params from GridSearchCV
final_model = XGBClassifier(
    use_label_encoder=False,
    eval_metric='logloss',
    n_estimators=200,
    max_depth=5,
    learning_rate=0.1
)

pipeline = Pipeline([
    ('scaler', RobustScaler()),
    ('clf', final_model)
])

pipeline.fit(X_res, y_res)

# Save model to file
joblib.dump(pipeline, 'xgboost_fraud_model.pkl')
print("✅ Model saved as xgboost_fraud_model.pkl")
```
